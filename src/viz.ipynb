```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
baseline_val = 98.48
baseline_test = 98.7

x = [25, 50, 95]
random_val = [98.49, 97.97, 92.66]
random_test = [98.61, 98.33, 93.26]

pruning_val = [98.60, 98.25, 91.10]
pruning_test = [98.63, 98.44, 92.23]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axes[0].axhline(y=baseline_val, color='black', linestyle='-', linewidth = 2)
axes[0].plot(x, random_val, color='blue', label='Random', marker='o')
axes[0].plot(x, pruning_val, color='red', label='Dyn-Unc', marker='x')
axes[0].set_title('Validation Set')
axes[0].set_xlabel('Pruning Ratio')
axes[0].set_ylabel('Acc. (%)')
axes[0].grid(True, linestyle='--', alpha=0.5)
axes[0].set_xticks(x)
axes[0].set_xticklabels(x)
axes[0].legend()

axes[1].axhline(y=baseline_test, color='black', linestyle='-', linewidth = 2)
axes[1].plot(x, random_test, color='blue', label='Random', marker='o')
axes[1].plot(x, pruning_test, color='red', label='Dyn-Unc', marker='x')
axes[1].set_title('Test Set')
axes[1].set_xlabel('Pruning Ratio')
axes[1].set_ylabel('Acc. (%)')
axes[1].grid(True, linestyle='--', alpha=0.5)
axes[0].set_xticks(x)
axes[0].set_xticklabels(x)
axes[1].legend()

plt.tight_layout()
plt.show()
```
